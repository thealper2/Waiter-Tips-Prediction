```python
import pandas as pd
import numpy as np
import plotly.express as px
import pickle
import plotly.graph_objects as go
from plotly.offline import init_notebook_mode
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import LabelEncoder
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv("tips.csv")
```

```python
fig = px.scatter(df, x="total_bill", y="tip", size="size", color="day", trendline="ols")
fig.show()
```

```python
def label_encoder(column):
    le = LabelEncoder().fit(column)
    print(column.name, le.classes_)
    return le.transform(column)
```

```python
columns = ["sex", "smoker", "day", "time"]
```

```python
# Female:0, Male:1
# No:0, Yes:1
# Fri:0, Sat:1, Sun:2, Thur:3
# Dinner:0, Lunch:1
for col in columns:
    df[col] = label_encoder(df[col])
```

```python
X = df.drop(["tip"], axis=1)
X = np.array(X)

y = df["tip"]
y = np.array(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=4242)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
test = [[16.99, 0, 0, 2, 0, 2]]
test = np.array(test)
model.predict(test)
```

```python
pickle.dump(model, open("model.pkl", "wb"))
```
